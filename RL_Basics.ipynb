```python
import numpy as np
grid_size = 5
agent_position = [0, 0]
goal_position = [4, 4]
print("Agent:", agent_position)
print("Goal:", goal_position)
```

```python
actions = {
    0: "UP",
    1: "DOWN",
    2: "LEFT",
    3: "RIGHT"
}
print(actions)
```

```python
def move_agent(position, action):
    x, y = position
    if action == 0 and x > 0:
        x -= 1
    elif action == 1 and x < 4:
        x += 1
    elif action == 2 and y > 0:
        y -= 1
    elif action == 3 and y < 4:
        y += 1
    return [x, y]
```

```python
agent_position = [0, 0]
new_position = move_agent(agent_position, 3)
print("New Position:", new_position)
```

```python
def get_reward(position, goal):
    if position == goal:
        return 10
    return -1
```

```python
reward = get_reward([4,4], goal_position)
print("Reward:", reward)
```

```python
import random
agent_position = [0, 0]
for step in range(10):
    action = random.choice([0,1,2,3])
    agent_position = move_agent(agent_position, action)
    reward = get_reward(agent_position, goal_position)
    print(f"Step {step+1}")
    print("Position:", agent_position)
    print("Reward:", reward)
    if reward == 10:
        print("Goal Reached!")
        break
```

```python
import random
episodes = 50
episode_rewards = []
for episode in range(episodes):
    agent_position = [0, 0]
    total_reward = 0
    for step in range(50):  # max steps per episode
        action = random.choice([0, 1, 2, 3])  # exploration
        agent_position = move_agent(agent_position, action)
        reward = get_reward(agent_position, goal_position)
        total_reward += reward
        if reward == 10:  # goal reached
            break
    episode_rewards.append(total_reward)
    import matplotlib.pyplot as plt 
    plt.plot(episode_rewards)
    plt.xlabel("Episode")
    plt.ylabel("Total Reward")
    plt.title("Reward Optimization Over Episodes")
    plt.show()
    print(f"Episode {episode+1}: Total Reward = {total_reward}")
```
